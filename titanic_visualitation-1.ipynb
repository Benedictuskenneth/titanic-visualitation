```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
titanic_data = pd.read_csv("/Users/kennethlukito/Downloads/train.csv")

# Display the first few rows
print(titanic_data.head())

# Check for missing values
print(titanic_data.info())
```

```python



# 1. Survival Count by Gender
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.countplot(x="Survived", hue="Sex", data=titanic_data, palette="coolwarm")
plt.title("Survival Count by Gender")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()

# 2. Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['Age'].dropna(), bins=30, kde=True, color="purple")
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()


```

```python
numeric_data = titanic_data.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_data.corr()
numeric_data = numeric_data.dropna()
numeric_data = numeric_data.fillna(numeric_data.mean())
correlation_matrix = numeric_data.corr()

# 3. Correlation Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix of Titanic Features")
plt.show()
```

```python
# 4. Fare by Class
plt.figure(figsize=(10, 6))
sns.boxplot(x="Pclass", y="Fare", data=titanic_data, palette="Set3")
plt.title("Fare by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()



```

```python

# 5. Pair Plot
selected_features = titanic_data[["Survived", "Pclass", "Age", "Fare"]]
sns.pairplot(selected_features, hue="Survived", palette="coolwarm")
plt.suptitle("Pair Plot of Titanic Features", y=1.02)
plt.show()
```
